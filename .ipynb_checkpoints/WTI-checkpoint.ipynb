```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
from pathlib import Path
%matplotlib inline

import os
import requests
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
```

```python
WTI_data = Path("./Resources/WTI.csv")
WTI_df = pd.read_csv(WTI_data, index_col="Day", parse_dates=True, infer_datetime_format=True)
WTI_df.sort_index(inplace=True)
WTI_df.head()
```

```python
WTI_df.columns=["WTI Spot Price"]
WTI_df.head()
```

```python
WTI_returns = WTI_df.pct_change().dropna()
WTI_returns.head()
```

```python
WTI_returns = WTI_returns.rename(columns={'WTI Spot Price': 'WTI_Returns'})
WTI_returns.head()
```

```python
WTI_returns.plot(figsize=(20,10), title = "Daily Returns")
```

```python
WTI_cumulative = (1+WTI_returns).cumprod()
WTI_cumulative.plot(figsize=(20,10), title="Cumulative Performance")
```

```python
MES_data = Path("./Resources/MES.csv")
MES_df = pd.read_csv(MES_data, index_col="Date", parse_dates=True, infer_datetime_format=True)
MES_df.sort_index(inplace=True)
MES_df.head()
```

```python
MES_df.dtypes
```

```python
MES_df["Close"] = MES_df["Close"].astype("float")
MES_df.drop(columns=["Open", "High", "Low", "Adj Close", "Volume"], inplace=True)
MES_df.head()
```

```python
MES_returns = MES_df.pct_change().dropna()
MES_returns.head()
```

```python
MES_returns = MES_returns.rename(columns={'Close': 'MES_Returns'})
MES_returns.head()
```

```python
ES_data = Path("./Resources/ES.csv")
ES_df = pd.read_csv(ES_data, index_col="Date", parse_dates=True, infer_datetime_format=True)
ES_df.sort_index(inplace=True)
ES_df.head()
```

```python
ES_df["Close/Last"] = ES_df["Close/Last"].astype("float")
ES_df.head()
```

```python
ES_df.drop(columns=["Volume", "Open", "High", "Low"], inplace=True)
```

```python
ES_df.head()
```

```python
ES_returns = ES_df.pct_change().dropna()
ES_returns = ES_returns.rename(columns={'Close/Last': 'ES_Returns'})
ES_returns.head()
```

```python
print(ES_returns.head())
print(MES_returns.head())
print(WTI_returns.head())
```

```python
load_dotenv()
```

```python
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")
```

```python
alpaca = tradeapi.REST(alpaca_api_key, alpaca_secret_key, api_version='v2')
```

```python
start = pd.Timestamp('2019-05-01', tz='America/New_York').isoformat()
end = pd.Timestamp('2023-05-01', tz='America/New_York').isoformat()
```

```python
tickers=["SPY"]
```

```python
timeframe="1Day"
```

```python
portfolio_df_year = alpaca.get_bars(tickers, timeframe, start=start, end=end).df
portfolio_df_year.head(10)
```

```python
SPY_closing = pd.DataFrame()
SPY_closing['close'] = portfolio_df_year['close']
SPY_closing.index = portfolio_df_year.index.date
SPY_closing.head()
```

```python
SPY_daily_return = SPY_closing.pct_change().dropna()
SPY_daily_return = SPY_daily_return.rename(columns={'close': 'SPY_returns'})
SPY_daily_return.index = pd.to_datetime(SPY_daily_return.index)
SPY_daily_return.head()
```

```python
ES_cumulative = (1+ES_returns).cumprod()
ES_cumulative.plot(figsize=(20,10), title="Cumulative Performance")
```

```python
MES_cumulative = (1+MES_returns).cumprod()
MES_cumulative.plot(figsize=(20,10), title="Cumulative Performance")
```

```python
SPY_cumulative = (1+SPY_daily_return).cumprod()
SPY_cumulative.plot(figsize=(20,10), title="Cumulative Performance")
```

```python
joined_returns = pd.concat([ES_returns, MES_returns, WTI_returns, SPY_daily_return], axis="columns", join="inner")
joined_returns.head()
```

```python
joined_returns.tail()
```

```python
joined_cumulative = (1+joined_returns).cumprod()
```

```python
joined_cumulative.plot(figsize=(20,10), title="Cumulative Performance")
```

```python
joined_returns_ex_WTI = pd.concat([ES_returns, MES_returns, SPY_daily_return], axis="columns", join="inner")
joined_returns_ex_WTI.head()
```

```python
joined_returns_cumulative_2 = (1+joined_returns_ex_WTI).cumprod()
joined_returns_cumulative_2.plot(figsize=(20,10), title="Cumulative Performance")
```

```python

correlation = joined_returns.corr()

mask=np.zeros_like(correlation)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(correlation, vmin=-1, vmax=1, annot=True, mask=mask)
```
