```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
from pathlib import Path
%matplotlib inline
```

```python
WTI_data = Path("./Resources/WTI.csv")
WTI_df = pd.read_csv(WTI_data, index_col="Day", parse_dates=True, infer_datetime_format=True)
WTI_df.sort_index(inplace=True)
WTI_df.head()
```

```python
WTI_df.columns=["WTI Spot Price"]
WTI_df.head()
```

```python
WTI_returns = WTI_df.pct_change().dropna()
WTI_returns.head()
```

```python
WTI_returns.plot(figsize=(20,10), title = "Daily Returns")
```

```python
WTI_cumulative = (1+WTI_returns).cumprod()
WTI_cumulative.plot(figsize=(20,10), title="Cumulative Performance")
```
